```python
from pathlib import Path
import re

input_file = Path(r"G:\\SteamLibrary\\steamapps\\common\\X-Plane 12\\Resources\\plugins\\Commands.txt")
output_file = Path(r"G:\\SteamLibrary\\steamapps\\common\\X-Plane 12\\Resources\\plugins\\Commands_Mod.txt")

separator = "|"
pattern = re.compile(r'^(sim/[^\s]+)\s+(.*)$')

with input_file.open("r", encoding="utf-8") as infile, output_file.open("w", encoding="utf-8") as outfile:
    for line in infile:
        line = line.strip()
        if not line:
            continue

        match = pattern.match(line)
        if match:
            col1, col2 = match.groups()
            outfile.write(f"{col1}{separator}{col2}\n")
        else:
            outfile.write(line + "\n")
print(f"File processed successfully and saved to: {output_file}")
```
